# Основы работы с библиотекой `matplotlib`

`Matplotlib` – одна из основных библиотек визуализации на Python.

Чтобы начать работу с `Matplotlib`, необходимо импортировать модуль `pyplot`, который содержит функции для работы с графиками:

```python
import matplotlib.pyplot as plt
```

Нарисуем простейший график:

```python
plt.plot([4, 2, 5, 4, 150])
```

В Jupyter Notebook график отображается автоматически, однако в обычном Python-скрипте следует вызывать функцию `show()`:

```python
plt.plot([69, 5, 3, 4, 5])
plt.show()
```

Видно, что значению 1 на оси `y` соответствует 0 на оси `x`. Дело в том, что `Matplotlib` воспринимает единственный список в функции `plot` как список значений `y`, а значения `x` генерируются автоматически, начиная с нуля.

```python
plt.plot(x, y, "b^-");
```

```python
plt.plot(x, y, "gD--");
```

При построении графиков следует иметь в виду, что функция `plot()` соединяет точки последовательно, в том порядке, в котором они следуют в списках или массивах. Из-за этой особенности, допустив ошибку, связанную с заданием неверной области определения функции, можно получить некорректные графики. Построим для примера гиперболу:

```python
x = np.linspace(-100, 100, 100)
y = 1/x
plt.plot(x, y);
```

Полученный график не является гиперболой: как известно, в точке $x=0$ график гиперболы уходит на бесконечность, линия при $x=0$ отсутствует. Избавимся от нее, построив график "по кусочкам":

```python
x1 = np.linspace(-100, -1, 100) # x < 0
y1 = 1/x1

x2 = np.linspace(1, 100, 100)   # x > 0
y2 = 1/x2

plt.plot(x1, y1, "blue")
plt.plot(x2, y2, "blue")
```

Добавим на график оси:

```python
x1 = np.linspace(-100, -1, 100) # x < 0
y1 = 1/x1

x2 = np.linspace(1, 100, 100)   # x > 0
y2 = 1/x2

plt.plot(x1, y1, "blue")
plt.plot(x2, y2, "blue")

plt.axhline(0, color='grey') # оси
plt.axvline(0, color='grey');
```

Основным объектом `Matplotlib` является Рисунок ([`Figure`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html)).  
`Figure` включает следующие элементы ([ссылка](https://matplotlib.org/stable/users/explain/quick_start.html#parts-of-a-figure)):

```python
import matplotlib.pyplot as plt
import numpy as np
```

Создадим объект `Figure` и область `Axes` (`subplot`), в которой будут отображаться данные.  
Параметр `1` у `figure` обозначает метку рисунка (`num`), с помощью которого можно на него ссылаться. Может быть строкой.   
`aspect` обозначает соотношение величин единиц на осях: `y/x`.

```python
fig = plt.figure(1)
ax = plt.subplot(aspect=1)
ax.plot(range(10))
plt.show()
```

Добавим вторую область `Axes`. Для этого используем параметр позиции в функции `subplot`. Этот параметр может задан следующим образом:
- три целых числа: `(nrows, ncols, index)`:
    - `nrows` – количество строк на рисунке;
    - `ncols` – количество индекс на рисунке;
    - `index` – позиция (индекс) данной области, начиная с левого верхнего угла;
- одно целое трехзначное число (в случае, если областей не более 9), где каждая цифра имеет тот же смысл, что в предыдущем варианте.

Например:
- `(2, 2, 1)` или `221` – на рисунке 2 строки, 2 столбца, данная область располагается в верхнем левом углу;
- `(2, 2, 3)` или `223` – на рисунке 2 строки, 2 столбца, данная область располагается в нижнем левом углу;
- `(3, 3, 5)` или `335` – на рисунке 3 строки, 3 столбца, данная область располагается посередине;

```python
fig = plt.figure(num=1, figsize=(12, 6))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)
ax1.plot(range(10))
ax2.plot(range(10))
plt.show()
```

Поэкспериментируйте с различными сочетаниями параметров в `subplot`, чтобы лучше понять принцип.

Изменим соотношение `aspect`:

```python
fig = plt.figure(figsize=(12, 6))
ax1 = plt.subplot(121, aspect=1)
ax2 = plt.subplot(122, aspect=0.5)
ax1.plot(range(10))
ax2.plot(range(10));
```

Можно сократить код с помощью функции `subplots`:

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, subplot_kw={"aspect": 1}, figsize=(12, 6))
ax1.plot(range(10))
ax4.plot(range(10));
```

Добавим линии сетки (`plt.grid`):

```python
x = np.linspace(-5, 5, 100)
y = x ** 2
z = x ** 3
r = np.exp(x)
m = abs(x)

fig = plt.figure(figsize=(10, 8))

plt.subplot(221)
plt.plot(x, y) # x^2
plt.title('Парабола')
plt.grid(True)

plt.subplot(222)
plt.plot(x, z) # x^3
plt.title('Кубическая функция')
plt.grid(True)

plt.subplot(223)
plt.plot(x, r) # e^x
plt.title('Экспонента')
plt.grid(True)

plt.subplot(224)
plt.plot(x, m) # |x|
plt.title('Модуль')
plt.grid(True)
```

Любая диаграмма включает следующие примитивы:
- графические элементы – маркеры, столбцы и т.п.;
- линии;
- текст.  

К каждому из этих примитивов можно получить доступ (например, к отметкам на осях):

```python
fig, ax = plt.subplots(figsize=(5,2))
for label in ax.get_xaxis().get_ticklabels():
    label.set_fontweight("bold")
```

Разрешение – количество точек на дюйм: dots per inch, dpi.  

По умолчанию в `Matplotlib` – `dpi=100` (но в Jupyter Notebook – `dpi=72`).  

`plt.figure(figsize=(6,6), dpi=100)` – рисунок 600x600 точек

```python
def plot(figsize, dpi):
    fig, ax = plt.subplots(figsize=figsize, dpi=dpi)
    ax.set_title(f"Figsize: {figsize}, dpi: {dpi}")
    ax.plot([2, 4, 1, 5], label="Label")
    ax.legend()

plot(figsize=(2, 2), dpi=200)
plot(figsize=(2, 2), dpi=100)
```

### Полярная система координат
За построение в определённой системе координат отвечает параметр `projection`.  
В `Matplotlib` поддерживаются следующие проекции (по умолчанию – `rectilinear`):

```python
import matplotlib
matplotlib.projections.get_projection_names()
```

Для построении графика в полярной системе координат нужно указать либо `projection='polar'`, либо `polar=True`.

Ниже приведён пример графика функции $5cos(3\phi + 4)$:

```python
plt.subplot(projection='polar')     # задаём полярную систему координат

phi = np.arange(0, 2*np.pi, 0.01)   # угол phi - массив от от 0 до 2*pi с шагом 0.01
ro = 5 * np.cos(3 * phi + 4)        # функция
plt.plot(phi, ro, linestyle='--');
```

Также в полярной системе кординат можно строить с помощью функции `polar()`:

```python
plt.polar(phi, 5*np.cos(3*phi + 4), linestyle='-', color='r');
```

### Цвета

Способы задания цвета:
- RGB или RGBA (alpha): `(0.5, 0.8, 0.2, 0.9)`
- hex RGB или RGBA: `#9467bd`, `#bcbd22fe`
- сокращенный hex RGB или RGBA: `#8bd = #88bbdd`
- одна буква: `"b"` = blue, `"k"` = black
- имена X11/CSS4 (регистронезависимые): `"aquamarine"`
- имена Tableau (палитра T10, регистронезависимые): `"tab:blue"`
- кортеж цвета и alpha: `("green", 0.3)`, `(“#f00", 0.3)`

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 6))
ax1.plot(range(10), color=(0.5, 0.8, 0.2, 0.9))
ax2.plot(range(10), color="#9467bd")
ax3.plot(range(10), color="r")
ax4.plot(range(10), color="aquamarine");
```

Нарисуем диаграмму рассеяния, задавая для каждой точки свой цвет (`colors = x`) и размер (`area = x**3`).  
При этом для преобразования числа в конкретный цвет испольуется `cmap` (по умолчанию `viridis`).

```python
plt.cm.cmaps_listed
```

```python
x = np.arange(0, 10, 0.1)
y = x**2
area = x**3
colors = x

plt.scatter(x, y, c=colors, s=area, cmap=plt.cm.magma)
plt.colorbar();                                        # добавление цветовой панели
```

```python
x = np.arange(0, 10, 0.4)
y = np.sin(x)
area = x**3
colors = x

plt.scatter(x, y, c=colors, s=area, cmap=plt.cm.Blues)
plt.colorbar();
```

Суффикс `_r` в названии карты означает обратный порядок цветов:

```python
plt.scatter(x, y, c=colors, s=area, cmap=plt.cm.Blues_r)
plt.colorbar();
```
