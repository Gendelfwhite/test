**1.** Пользователь вводит строку, символ (букву) и подстроку. В исходной строке после каждой заданной пользователем буквы вставьте заданную подстроку.

```python
input_string = input("Введите строку: ")
input_char = input("Введите символ: ")
input_substring = input("Введите подстроку: ")

# Инициализация результирующей строки
result = ""

# Цикл по строке
for i in range(len(input_string)):
    result += input_string[i]  # Добавляем текущий символ
    if input_string[i] == input_char:
        result += input_substring  # Если символ совпадает, добавляем подстроку

# Вывод результата
print("Измененная строка:", result)

import re

input_string = input("Введите строку: ")
input_char = input("Введите символ: ")
input_substring = input("Введите подстроку: ")

# Регулярное выражение для поиска всех вхождений символа
symbol = re.escape(input_char)

# Замена каждого вхождения символа на символ + подстрока
result = re.sub(symbol, input_char + input_substring, input_string)

# Вывод результата
print("Измененная строка:", result)
```

**2.** Подсчитайте сумму целых чисел, встречающихся в строке. Символ ‘–’ перед числом считается знаком отрицательного числа. Все остальные символы (кроме цифр и минуса) считаются разделителями между числами.

```python
input_string = input("Введите строку: ")

total_sum = 0
current_number = ''
is_negative = False

# Перебираем каждый символ в строке
for char in input_string:
    if char == '-':
        is_negative = True
    elif char.isdigit():
        current_number += char
    else:
        if current_number:
            if is_negative:
                total_sum -= int(current_number)
                is_negative = False
            else:
                total_sum += int(current_number)
            current_number = ''

# Обработка последнего числа (если оно есть)
if current_number:
    if is_negative:
        total_sum -= int(current_number)
    else:
        total_sum += int(current_number)

# Вывод результата
print("Сумма чисел в строке:", total_sum)

import re

input_string = input("Введите строку: ")

# Регулярное выражение для поиска всех целых чисел (положительных и отрицательных)
pattern = r'-?\d+'

# Найдем все числа в строке
numbers = re.findall(pattern, input_string)

# Преобразуем найденные строки в числа и подсчитаем сумму
total_sum = sum(int(num) for num in numbers)

# Вывод результата
print("Сумма чисел в строке:", total_sum)
```

**3.** Сформируйте новую строку, отличающуюся от исходной тем, что группы идущих подряд одинаковых символов разделены символом \*. Пример ееннне -> ее\*ннн\*е, ппрол-> пп\*р\*о\*л.

```python
input_string = input("Введите строку: ")
result = input_string[0]  # Начинаем с первого символа строки

# Перебираем каждый символ, начиная со второго
for i in range(1, len(input_string)):
    if input_string[i] == input_string[i - 1]:
        result += input_string[i]  # Добавляем символ, если он такой же, как предыдущий
    else:
        result += '*' + input_string[i]  # Добавляем '*' и новый символ, если он отличается

# Вывод результата
print("Измененная строка:", result)

import re
input_string = input("Введите строку: ")

# Регулярное выражение для поиска групп одинаковых символов
pattern = r'(.)\1*'

# Замена каждой группы на саму группу с добавлением '*' после нее
result = ''
i = 0
while i < len(input_string):
    count = 1
    while i + 1 < len(input_string) and input_string[i] == input_string[i+1]:
        i += 1
        count += 1
    result += input_string[i] * count + '*'
    i += 1

# Удаление лишних '*' в конце строки
result = result.rstrip('*')

# Вывод результата
print("Измененная строка:", result)
```

**5.** Преобразовать сегодняшнюю дату в системном представлении (2024-09-10) в привычный для нас формат день/месяц/год (17/09/2018). Известно, что на год выделено всегда 4 цифры, а на день и месяц – всегда 2 цифры.

Примечание:
за работу с датами и временем в Python отвечает модуль datetime, а непосредственно для работы с датами используется объект date и его методы.

```python
from datetime import datetime

system_date = "2024-09-10"
date_obj = datetime.strptime(system_date, "%Y-%m-%d")
formatted_date = date_obj.strftime("%d/%m/%Y")
print("Преобразованная дата:", formatted_date)

system_date = "2024-09-10"
year, month, day = system_date.split("-")
formatted_date = f"{day}/{month}/{year}"
print("Преобразованная дата:", formatted_date)
```

**6.** Выведите слова сообщения, являющиеся палиндромами, по убыванию их длин. Для повторяющихся слов в скобках указывать количество повторов.

```python
message = "Мадам, я видел радар, а вы видели мадам?"
words = message.lower().split()

palindromes = {}
for word in words:
    word = ''.join(e for e in word if e.isalnum())
    if word == word[::-1]:
        if word in palindromes:
            palindromes[word] += 1
        else:
            palindromes[word] = 1

for word, count in sorted(palindromes.items(), key=lambda x: len(x[0]), reverse=True):
    if count > 1:
        print(f"{word} ({count})")
    else:
        print(word)


import re

message = "Мадам, я видел радар, а вы видели мадам?"
words = re.findall(r'\b\w+\b', message.lower())

palindromes = {}
for word in words:
    if word == word[::-1]:
        if word in palindromes:
            palindromes[word] += 1
        else:
            palindromes[word] = 1

for word, count in sorted(palindromes.items(), key=lambda x: len(x[0]), reverse=True):
    if count > 1:
        print(f"{word} ({count})")
    else:
        print(word)
```
