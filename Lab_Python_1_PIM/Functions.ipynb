# Функции в Python


Синтаксис

def ИМЯ_ФУНКЦИИ(параметры):

    выражение № 1
    
    выражение № 2 ...


Вызов:

ИМЯ_ФУНКЦИИ(параметры)

```python
def func(a):
    print(a)
    a += 1

a = 5
func(a)
print(a)
```

```python
def func(a):
    print(a)
    a += 1
    return a

a = 5
print(func(a))
```

```python
def sum(a,b):
    return a+b

print(sum(2,3))
print(sum('12','34'))
```

```python
def func(a, b):
    print(a, b)
    a += 1
    b -= 1
    return a, b

a = 5
b = 5.0
a, b = func(a, b)
print(a, b)
print(func(a, b))
print(a, b)
```

```python
def divide(delimoe, delitel):
    return delimoe // delitel

print(divide(50, 5))
print(divide(delitel = 5, delimoe = 50))
```

```python
def divide(delimoe, delitel=2):
    return delimoe // delitel

print(divide(50, 5))
print(divide(50))
print(divide(delimoe = 50))
print(divide(delitel = 5, delimoe = 50))
```

```python
def example(a):
    if a>0:
        return "positive"
    else:
        a =- a
        return a

print(example(5))
print(example(-5))
```

```python
def rec(a, b):
    if b > 1:
        return a * rec(a, b - 1)
    else:

        return a

print(rec(2, 10))
print(rec(2, 600))
print(rec(2, 900))
```

```python
### lambda arguments: expression

double = lambda x: x*2
print(double(35))
```

```python
func = lambda x, y: x+y

print(func(2, 3))
print(func('a', 'b'))
```

```python
print((lambda x, y: x+y)(2,3))
print((lambda x, y: x+y)('a','b'))
```

```python
func = lambda *args: args
print(func(1, 3, 5, 7))
```

```python
func = lambda *args: args*2
print(func(1, 3, 5, 7))
```
