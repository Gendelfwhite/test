5) Добавьте столбцы FirstNameList с именем пассажира. Обратите внимание, что встречается разный формат хранения имен, в том числе существуют отличия для мужчин и женщин. Имя может быть составное (из нескольких имен) - рассмотрите оба варианта - составное имя целиком и составное имя как отдельные имена. Какое самое популярное женское имя на корабле? Какое самое популярное мужское имя на корабле? У замужних женщин имя берём в скобках

```python
import pandas as pd
import re
df_titanic["Appeal"] = df_titanic["Name"].str.extract(r",\s*([A-Za-z.]+\s*[A-Za-z]+)\.")
df_titanic
```

```python
# Разделение имени
split_names = df_titanic["Name"].str.split(",", n=1, expand=True)

# Извлечение имени для "Mrs"
def extract_name(name):
    # Попытка извлечь имя в скобках
    name_in_brackets = re.search(r"\s*\((.*?)\)", name)
    if name_in_brackets:
        return name_in_brackets.group(1).strip()
    # Если в скобках нет, попытка извлечь имя после точки
    name_after_dot = re.search(r"\.\s*([A-Za-z.]+)", name)
    if name_after_dot:
        return name_after_dot.group(1).strip()
    return None

# Применение функции к строкам с "Mrs"
df_titanic.loc[df_titanic["Appeal"] == "Mrs", "FirstNameList"] = split_names.loc[df_titanic["Appeal"] == "Mrs", 1].apply(extract_name)

# Извлечение имени для других значений "Appeal"
df_titanic.loc[df_titanic["Appeal"] != "Mrs", "FirstNameList"] = split_names.loc[df_titanic["Appeal"] != "Mrs", 1].str.split(".", n=1, expand=True)[1].str.strip()

# Удаление лишних пробелов и символов
df_titanic["FirstNameList"] = df_titanic["FirstNameList"].str.replace(r'[^A-Za-z\s]', '', regex=True).str.strip()

# Удаление слов длиной 2 и менее символов
df_titanic["FirstNameList"] = df_titanic["FirstNameList"].apply(lambda x: ' '.join([word for word in x.split() if len(word) > 2]) if pd.notnull(x) else x)

df_titanic.head(20)

```

```python
# Разделение на женские и мужские данные с использованием столбца "Sex"
female_names = df_titanic[df_titanic["Sex"] == "female"]["FirstNameList"]
male_names = df_titanic[df_titanic["Sex"] == "male"]["FirstNameList"]

# Функция для подсчета частот имен
def count_names(names):
    # Разделение столбца на отдельные имена
    all_names = names.dropna().str.split(expand=True).stack()
    # Подсчет частоты имен
    name_counts = all_names.value_counts()
    return name_counts

# Подсчет частот имен для женщин и мужчин
female_name_counts = count_names(female_names)
male_name_counts = count_names(male_names)

# Получение самого популярного женского и мужского имени
most_popular_female_name = female_name_counts.idxmax()
most_popular_male_name = male_name_counts.idxmax()

# Подсчет частот составных имен
def count_full_names(names):
    # Подсчет частот составных имен
    full_name_counts = names.dropna().value_counts()
    return full_name_counts

# Подсчет частот составных имен для женщин и мужчин
female_full_name_counts = count_full_names(female_names)
male_full_name_counts = count_full_names(male_names)

# Получение самого популярного составного имени для женщин и мужчин
most_popular_female_full_name = female_full_name_counts.idxmax()
most_popular_male_full_name = male_full_name_counts.idxmax()

# Результаты
print("Самое популярное женское имя (как отдельные имена):", most_popular_female_name)
print("Самое популярное женское составное имя:", most_popular_female_full_name)
print("Самое популярное мужское имя (как отдельные имена):", most_popular_male_name)
print("Самое популярное мужское составное имя:", most_popular_male_full_name)
```

6) Приведите список уникальных (неповторяющихся, только один человек имеет это имя) имен пассажиров, отдельно мужских и женских. Рассмотрите оба варианта - составное имя целиком и составное имя как отдельные имена. Выведите таблицу с этими пассажирами.

```python
# Функция для проверки уникальности имени
def is_unique_name(name, sex, df):
    count = len(df[(df['FirstNameList'] == name) & (df['Sex'] == sex)])
    return count == 1

# Разделение имен на составные и отдельные
df_titanic['IsCompoundName'] = df_titanic['FirstNameList'].apply(lambda x: len(x.split()) > 1 if pd.notnull(x) else False)

# Нахождение уникальных имен
unique_compound_names = []
unique_single_names = []

for name, row in df_titanic.iterrows():
    if pd.notnull(row['FirstNameList']):
        if is_unique_name(row['FirstNameList'], row['Sex'], df_titanic):
            if row['IsCompoundName']:
                unique_compound_names.append({'Пол': row['Sex'], 'Имя': row['FirstNameList']})
            else:
                unique_single_names.append({'Пол': row['Sex'], 'Имя': row['FirstNameList']})

# Создание таблицы для составных имен
compound_names_df = pd.DataFrame(unique_compound_names)

# Создание таблицы для отдельных имен
single_names_df = pd.DataFrame(unique_single_names)

# Вывод таблиц
print("Таблица составных имен:")
print(compound_names_df)
print("\nТаблица отдельных имен:")
print(single_names_df)
```

7) Приведите множество всех имен пассажиров, отдельно мужских и женских. Рассмотрите оба варианта - составное имя целиком и составное имя как отдельные имена.

```python
# Разделение составных имен на отдельные
df_titanic["SingleNames"] = df_titanic["FirstNameList"].apply(lambda x: x.split() if pd.notnull(x) and len(x.split()) > 1 else [x])

# Создание множеств имен
male_compound_names = set(df_titanic[df_titanic['Sex'] == 'male']['FirstNameList'].dropna())
female_compound_names = set(df_titanic[df_titanic['Sex'] == 'female']['FirstNameList'].dropna())

male_single_names = set()
female_single_names = set()

for index, row in df_titanic.iterrows():
    if pd.notnull(row['FirstNameList']):
        for name in row['SingleNames']:
            if row['Sex'] == 'male':
                male_single_names.add(name)
            elif row['Sex'] == 'female':
                female_single_names.add(name)

# Вывод множеств
print("Мужские составные имена (целиком):")
print(male_compound_names)
print("\nМужские отдельные имена:")
print(male_single_names)

print("\nЖенские составные имена (целиком):")
print(female_compound_names)
print("\nЖенские отдельные имена:")
print(female_single_names)
```

8) Приведите список имен, которые одновременно встречаются у мужчин, и у женщин. Решите задачу, если рассматривать составные имена как одно имя, и если рассматривать составные имена отдельно. Выведите таблицу пассажирами, имеющими эти имена.

```python
# Разделение составных имен на отдельные
df_titanic["SingleNames"] = df_titanic["FirstNameList"].apply(lambda x: x.split() if pd.notnull(x) and len(x.split()) > 1 else [x])

# Создание множеств имен для мужчин и женщин
male_names = set(df_titanic[df_titanic['Sex'] == 'male']['FirstNameList'].dropna())
female_names = set(df_titanic[df_titanic['Sex'] == 'female']['FirstNameList'].dropna())

# Имена, встречающиеся у мужчин и женщин (составные как одно имя)
common_compound_names = male_names.intersection(female_names)

# Имена, встречающиеся у мужчин и женщин (составные как отдельные имена)
male_single_names = set()
female_single_names = set()

for index, row in df_titanic.iterrows():
    if pd.notnull(row['FirstNameList']):
        for name in row['SingleNames']:
            if row['Sex'] == 'male':
                male_single_names.add(name)
            elif row['Sex'] == 'female':
                female_single_names.add(name)

common_single_names = male_single_names.intersection(female_single_names)

# Фильтрация пассажиров с общими именами
common_compound_passengers = df_titanic[df_titanic['FirstNameList'].isin(common_compound_names)]
common_single_passengers = df_titanic[df_titanic['SingleNames'].apply(lambda x: any(name in common_single_names for name in x))]

# Вывод таблиц
print("Пассажиры с общими именами (составные как одно имя):")
print(common_compound_passengers[['Name', 'Sex', 'FirstNameList']])

print("\nПассажиры с общими именами (составные как отдельные имена):")
print(common_single_passengers[['Name', 'Sex', 'FirstNameList']])
```
