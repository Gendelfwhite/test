**1.** Разработайте программу на языке программирования Python, описывающую музыкальные записи в домашней коллекции. Любая запись характеризуется названием песни, исполнителем, названием альбома, годом выхода, длительностью. Предусмотреть возможность добавления записи в коллекцию, удаления записи, изменения данных о записи, вывода всех записей из указанного альбома определенного исполнителя, поиска записи по названию или длительности.

```python
class MusicRecord:
    def __init__(self, title, artist, album, year, duration):
        self.title = title
        self.artist = artist
        self.album = album
        self.year = year
        self.duration = duration

class MusicCollection:
    def __init__(self):
        self.records = []

    def add_record(self, record):
        self.records.append(record)

    def remove_record(self, title):
        for record in self.records:
            if record.title == title:
                self.records.remove(record)

    def update_record(self, title, new_record):
        for i, record in enumerate(self.records):
            if record.title == title:
                self.records[i] = new_record

    def get_records_by_album(self, artist, album):
        return [record for record in self.records if record.artist == artist and record.album == album]

    def search_record_by_title(self, title):
        return [record for record in self.records if record.title == title]

    def search_record_by_duration(self, duration):
        return [record for record in self.records if record.duration == duration]

# Пример использования
collection = MusicCollection()

record1 = MusicRecord("Song 1", "Artist 1", "Album 1", 2020, "3:30")
record2 = MusicRecord("Song 2", "Artist 1", "Album 1", 2020, "4:00")

collection.add_record(record1)
collection.add_record(record2)

print("All records in Album 1 by Artist 1:")
for record in collection.get_records_by_album("Artist 1", "Album 1"):
    print(f"{record.title} by {record.artist}")

print("\nSearching for 'Song 1':")
for record in collection.search_record_by_title("Song 1"):
    print(f"{record.title} by {record.artist}")

print("\nSearching for duration '4:00':")
for record in collection.search_record_by_duration("4:00"):
    print(f"{record.title} by {record.artist}")

collection.remove_record("Song 2")
print("\nAfter removing 'Song 2':")
for record in collection.records:
    print(f"{record.title} by {record.artist}")

"""Этот код предоставляет базовую реализацию системы управления музыкальными
записями в домашней коллекции на языке Python.
Можно дополнить его дополнительными функциями и улучшениями по своему усмотрению."""
```

```python
class Property:
    def __init__(self, address, size, rooms, rent):
        self.address = address
        self.size = size
        self.rooms = rooms
        self.rent = rent

    def display_info(self):
        print(f"Address: {self.address}")
        print(f"Size: {self.size} sq. ft")
        print(f"Number of rooms: {self.rooms}")
        print(f"Rent: ${self.rent}/month")

class RentalAgency:
    def __init__(self):
        self.properties = []

    def add_property(self, property):
        self.properties.append(property)

    def find_property_by_criteria(self, min_size, min_rooms, max_rent):
        filtered_properties = [prop for prop in self.properties
                               if prop.size >= min_size
                               and prop.rooms >= min_rooms
                               and prop.rent <= max_rent]
        return filtered_properties

# Пример использования
property1 = Property("123 Main St", 800, 2, 1200)
property2 = Property("456 Elm St", 1000, 3, 1500)

agency = RentalAgency()
agency.add_property(property1)
agency.add_property(property2)

print("Available properties matching criteria:")
filtered_properties = agency.find_property_by_criteria(900, 2, 1400)
for prop in filtered_properties:
    prop.display_info()

"""Этот пример программы демонстрирует простую реализацию информационной системы
жилищного агентства с использованием объектно-ориентированного подхода в Python.
Можно расширить этот код, добавив дополнительные функциональности и методы в соответствии с требованиями  проекта."""


```
