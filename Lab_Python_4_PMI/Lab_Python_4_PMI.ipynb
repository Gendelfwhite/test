**input.txt**

*apple – лесть, яблоко  
soap – лесть, мыло  
foam – мыло, пена*
  
**output.txt**

*лесть – apple, soap  
мыло – foam, soap  
пена – foam  
яблоко – apple*

```python
# Чтение данных из файла input.txt
with open('/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/input.txt', 'r', encoding='utf-8') as file:
  lines = file.readlines() #считывает все строки файла и возвращает их в виде списка

# Создание словаря для русско-английского перевода
ru_en_dict = {}

# Обработка каждой строки из файла
for line in lines:
  # Удаляем лишние пробелы и пропускаем пустые строки
  line = line.strip()
  if not line:
    continue

    # Проверяем, есть ли разделитель ' – ' в строке
  if ' – ' not in line:
    continue

    # Разделение строки на английское слово и список русских слов
  en_word, ru_words = line.split(' – ')
  ru_words_list = ru_words.split(', ')

    # Добавление каждого русского слова в словарь
  for ru_word in ru_words_list:
    if ru_word in ru_en_dict:
      ru_en_dict[ru_word].append(en_word)
    else:
      ru_en_dict[ru_word] = [en_word]

# Сортировка словаря по ключам (русским словам)
sorted_ru_en_dict = dict(sorted(ru_en_dict.items()))
print(sorted_ru_en_dict)
# Запись отсортированного словаря в файл output.txt
with open('output.txt', 'w', encoding='utf-8') as file:
  # Преобразуем словарь в список кортежей для удобства
  items = list(sorted_ru_en_dict.items())

  # Записываем все строки, кроме последней
  for ru_word, en_words in items[:-1]:
    file.write(f"{ru_word} – {', '.join(en_words)}\n")

  if items:
    ru_word, en_words = items[-1]
    file.write(f"{ru_word} – {', '.join(en_words)}")
```

#### Имя

Сохранить все данные в словарь (ключи – имена и фамилии пользователей, значения – все остальное).   
Затем программа случайным образом загадывает любую категорию (кроме имени и фамилии), выводит ее значение и предлагает угадать, что это был за друг. Если игрок не угадывает – предлагается другая категория. Если игрок не угадал по всем категориям, программа выдает правильный ответ.
После угадывания или неугадывания одного друга программа предлагает загадать другого друга или завершить выполнение.

```python
from google.colab import drive
import os
import random


# Путь к папке с файлами анкет
directory = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/friends_catalog'

friends_catalog = {}

# Чтение данных из всех файлов в указанной директории
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(directory, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
            # Извлечение данных по заданным индексам (при данной структуре файла)
            name = lines[1].strip()
            surname = lines[4].strip()
            full_name = f"{name} {surname}"

            age = lines[7].strip()
            favorite_food = lines[10].strip()
            music_band = lines[13].strip()
            dream = lines[16].strip()

            friends_catalog[full_name] = {
                "Возраст": age,
                "Любимая еда": favorite_food,
                "Музыкальная группа": music_band,
                "Заветная мечта": dream
            }

print("Добро пожаловать в игру 'Знаешь ли ты своих друзей?'!")

# Основной игровой цикл
while True:
    # Случайно выбираем друга
    friend_name, friend_info = random.choice(list(friends_catalog.items()))

    # Получаем список категорий и перемешиваем их
    categories = list(friend_info.keys())
    random.shuffle(categories)

    print("\nПопробуйте угадать, кто это!")
    guessed = False

    for category in categories:
        value = friend_info[category]
        print(f"\nКатегория: {category}")
        print(f"Подсказка: {value}")

        answer = input("Кто это? (Введите имя и фамилию): ").strip()
        # Сравнение ответов без учёта регистра
        if answer.lower() == friend_name.lower():
            print("Правильно! Вы угадали!")
            guessed = True
            break
        else:
            print("Неверно. Попробуйте еще раз.")

    if not guessed:
        print(f"\nВы не смогли угадать. Это был(а) {friend_name}.")

    # Предлагаем игроку продолжить игру
    choice = input("\nХотите попробовать снова? (да/нет): ").strip().lower()
    if choice != 'да':
        print("Спасибо за игру! До свидания!")
        break
```

**3.**	Создайте словарь, где ключами являются числа, а значениями – строки. Примените к нему метод items(), полученный объект dict_items передайте в написанную вами функцию, которая создает и возвращает новый словарь, обратный исходному, т.е. ключами являются строки, а значениями – числа (если соответствует только одно значение) или списки чисел (если несколько).

```python
def reverse_dict(items):
    """
    Функция для создания обратного словаря.

    :param items: Объект dict_items из исходного словаря
    :return: Обратный словарь
    """
    reversed_dict = {}

    for key, value in items:
        if value not in reversed_dict:
            # Если значение (строка) еще не встречалось, добавляем его как ключ
            reversed_dict[value] = [key]
        else:
            # Если значение уже есть, добавляем текущий ключ в список
            reversed_dict[value].append(key)

    # Преобразуем списки из одного элемента в простые значения
    for key, value in reversed_dict.items():
        if len(value) == 1:
            reversed_dict[key] = value[0]

    return reversed_dict


# Создаем исходный словарь
original_dict = {
    1: "apple",
    2: "banana",
    3: "apple",
    4: "cherry",
    5: "banana",
    6: "apple"
}

# Получаем объект dict_items
items = original_dict.items()

# Вызываем функцию для создания обратного словаря
reversed_dict = reverse_dict(items)

# Выводим результаты
print("Исходный словарь:")
print(original_dict)
print("\nОбратный словарь:")
print(reversed_dict)
```

**4.**	Имеется каталог X, содержащий файлы и другие каталоги. Создать копию структуры каталогов, где каждый файл *.txt обрабатывается следующим образом: первой строкой вставляется число строк в исходном файле, а последней строкой – уровень вложенности в каталог X. Остальные файлы пропускаются. Сам каталог X и его содержимое изменяться не должны.

```python
import os
import shutil

# Путь к исходному каталогу X на Google Диске
source_dir = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/X'

# Путь к копии структуры каталогов
target_dir = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/copy_of_X'

# Функция для определения уровня вложенности
def get_nesting_level(path, base_dir):
    # Уровень вложенности — количество разделителей в пути относительно базового каталога
    relative_path = os.path.relpath(path, base_dir)
    return relative_path.count(os.sep)

# Функция для рекурсивного копирования и обработки файлов
def copy_and_process_directory(source, target):
    # Если целевой каталог не существует, создаем его
    if not os.path.exists(target):
        os.makedirs(target)

    # Обходим содержимое исходного каталога
    for item in os.listdir(source):
        source_item = os.path.join(source, item)
        target_item = os.path.join(target, item)

        if os.path.isdir(source_item):
            # Если это каталог, рекурсивно вызываем функцию для него
            copy_and_process_directory(source_item, target_item)
        elif source_item.endswith('.txt'):
            # Если это .txt-файл, обрабатываем его
            process_txt_file(source_item, target_item)
        else:
            # Для всех остальных файлов просто копируем их
            shutil.copy2(source_item, target_item)

# Функция для обработки .txt-файла
def process_txt_file(source_file, target_file):
    # Подсчитываем количество строк в исходном файле
    with open(source_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        num_lines = len(lines)

    # Определяем уровень вложенности
    nesting_level = get_nesting_level(source_file, source_dir)

    # Создаем новый файл с обработанным содержимым
    with open(target_file, 'w', encoding='utf-8') as f:
        # Первая строка: количество строк
        f.write(f"{num_lines}\n")

        # Остальные строки из исходного файла
        f.writelines(lines)

        # Последняя строка: уровень вложенности
        f.write(f"{nesting_level}\n")

# Создаем копию структуры каталогов
copy_and_process_directory(source_dir, target_dir)

print(f"Копия структуры каталогов создана в {target_dir}")
```

**5.** Напишите программу, которая для заданного каталога создает подкаталоги, соответствующие дате создания каждого отдельного файла, и перемещает каждый файл в соответствующий дате каталог.

```python
import os
import shutil
from datetime import datetime

def organize_files(source_dir):
    """
    Перемещаем файлы в папки по дате создания
    """
    # Получаем список всех файлов в папке
    files = os.listdir(source_dir)

    # Перебираем каждый файл
    for file in files:
        # Пропускаем папки, работаем только с файлами
        file_path = os.path.join(source_dir, file)
        if os.path.isfile(file_path):
            # Получаем время создания файла
            creation_time = os.path.getctime(file_path)
            # Преобразуем время в дату (ГГГГ-ММ-ДД)
            date = datetime.fromtimestamp(creation_time).strftime('%Y-%m-%d')

            # Создаём папку с именем даты, если её нет
            date_folder = os.path.join(source_dir, date)
            if not os.path.exists(date_folder):
                os.makedirs(date_folder)

            # Перемещаем файл в папку с датой
            new_path = os.path.join(date_folder, file)
            shutil.move(file_path, new_path)
            print(f"Файл '{file}' перемещён в папку '{date}'")


folder_path = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/5_задание'

# Запускаем программу
print("Начинаем сортировку файлов...")
organize_files(folder_path)
print("Сортировка завершена!")
```

**6.**	Из файлов *shop1.txt* и *shop2.txt* считываются прайс-листы (наименование товара – цена). Необходимо объединить два прайс-листа так, чтобы если наименование товара присутствует в обоих листах, то в итоговый прайс-лист помещается большая цена. Результат записать в файл *shop_max.txt*.

```python
import re

shop1_path = "/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/6_задание/shop1.txt"
shop2_path = "/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/6_задание/shop2.txt"
output_path = "/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/6_задание/shop_max.txt"


def parse_price_list(file_path):
    """Функция для извлечения товаров и цен из файла"""
    price_dict = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            line = re.sub(r'^\d+\.\s+', '', line)  # Убираем номер строки

            # Ищем название товара и цену
            match = re.search(r'(.+?)\s+\.+\s+(\d+)\s+руб\.', line)
            if match:
                item = match.group(1).strip()  # Название товара
                price = int(match.group(2))  # Цена
                price_dict[item] = price
    return price_dict

# Считываем данные из обоих прайс-листов
shop1_prices = parse_price_list(shop1_path)
shop2_prices = parse_price_list(shop2_path)

# Проверяем, есть ли данные
if not shop1_prices and not shop2_prices:
    print("Ошибка: Оба файла пустые или формат неверный!")
else:
    # Объединяем прайс-листы, оставляя максимальную цену
    merged_prices = shop1_prices.copy()
    for item, price in shop2_prices.items():
        merged_prices[item] = max(merged_prices.get(item, 0), price)

    # Вычисляем максимальную длину названия товара для красивого выравнивания
    max_item_length = max(len(item) for item in merged_prices.keys())

    # Записываем результат в файл shop_max.txt
    with open(output_path, 'w', encoding='utf-8') as file:
        file.write("======================================\n")
        file.write("      ОБЪЕДИНЕННЫЙ ПРАЙС-ЛИСТ\n")
        file.write("======================================\n\n")

        for i, (item, price) in enumerate(sorted(merged_prices.items()), start=1):
            dots = '.' * (max_item_length - len(item) + 10)  # Ровняем точки
            file.write(f"{i}. {item} {dots} {price} руб.\n")

    print("Файл 'shop_max.txt' успешно создан!")
```

**7.** Создайте файл *input.txt*, содержащий данные о странах экспортерах (не менее 10): название страны, название экспортируемого товара и объем экспортируемого товара (в тоннах). Напишите программу, позволяющую обрабатывать следующие запросы пользователя:
- добавить данные в файл;

- вывести всю информацию, содержащуюся в файле;

- вывести список товаров, экспортируемых введенной пользователем страной;

- вывести список стран, экспортирующих заданный товар.

Если добавляется информация о уже имеющихся в базе стране и товаре, новое значение тонн должно прибавлять к уже имеющемуся количеству.

```python
file_path = "/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/7_задание/input.txt"

def read_data():
    data = {}
    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            country, product, quantity = line.strip().split("; ")
            quantity = int(quantity)
            if country not in data:
                data[country] = {}
            if product in data[country]:
                data[country][product] += quantity  # Суммируем тонны
            else:
                data[country][product] = quantity
    return data

# Функция для записи данных в файл
def write_data(data):
    with open(file_path, "w", encoding="utf-8") as file:
        for country, products in data.items():
            for product, quantity in products.items():
                file.write(f"{country}; {product}; {quantity}\n")

# Функция для вывода всех данных
def show_all():
    data = read_data()
    print("\n ВЕСЬ СПИСОК ЭКСПОРТА:")
    for country, products in data.items():
        for product, quantity in products.items():
            print(f"{country}: {product} - {quantity} тонн")

# Функция для добавления данных
def add_data():
    country = input("Введите страну: ").strip()
    product = input("Введите экспортируемый товар: ").strip()
    quantity = int(input("Введите объем экспорта (тонн): "))

    data = read_data()

    # Если страна уже есть, обновляем данные
    if country in data:
        if product in data[country]:
            data[country][product] += quantity
        else:
            data[country][product] = quantity
    else:
        data[country] = {product: quantity}

    write_data(data)
    print(f" Добавлено: {country} - {product} ({quantity} тонн)")

# Функция для поиска товаров по стране
def get_products_by_country():
    country = input("Введите название страны: ").strip()
    data = read_data()

    if country in data:
        print(f"\n {country} экспортирует:")
        for product, quantity in data[country].items():
            print(f"- {product}: {quantity} тонн")
    else:
        print(" Страна не найдена.")

# Функция для поиска стран по товару
def get_countries_by_product():
    product = input("Введите название товара: ").strip()
    data = read_data()
    countries = []

    for country, products in data.items():
        if product in products:
            countries.append(f"{country} ({products[product]} тонн)")

    if countries:
        print(f"\n {product} экспортируют:")
        for country in countries:
            print(f"- {country}")
    else:
        print(" Товар не найден.")

def menu():
    while True:
        print("\n Меню:")
        print("1. Вывести всю информацию")
        print("2. Добавить данные")
        print("3. Найти товары по стране")
        print("4. Найти страны по товару")
        print("5. Выход")

        choice = input("Выберите действие (1-5): ")

        if choice == "1":
            show_all()
        elif choice == "2":
            add_data()
        elif choice == "3":
            get_products_by_country()
        elif choice == "4":
            get_countries_by_product()
        elif choice == "5":
            print("Выход из программы.")
            break
        else:
            print("Неверный ввод! Попробуйте снова.")

# Запуск меню
menu()
```
