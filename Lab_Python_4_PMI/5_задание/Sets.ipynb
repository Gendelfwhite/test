# Множества

неотсортированная коллекция уникальных элементов

поддерживают итерацию, добавление и удаление объектов и т.д.

 индексация и срезы в сетах не поддерживаются

```python
s1 = set('abacaaaade')
s2 = set('aghij')
print(s1)
print(s2)
```

```python
s3 = s1 - s2
s3
```

```python
s3 = s1 | s2
s3
```

```python
s3 = s1 & s2
s3
```

```python
s1.add('6')
s1
```

```python
s1.remove('6')
s1
```

```python
for item in s1:
    print(item)
```

```python
L = [7, 7, 3, 4, 1, 2, 6, 3]
set(L)
```

```python
L = sorted(list(set(L)))
L
```

```python
div2 = [x * 2 for x in range(0, 11)]
div3 = [x * 3 for x in range(0, 7)]

div2_set = set(div2)
div3_set = set(div3)

div2_set & div3_set

```

```python
div2_set | div3_set
```

```python
div2_set - div3_set
```
